```python
df.columns = df.columns.str.strip() 
```

```python
feature_to_plot = "PopulationIndia"
```

```python
plt.figure(figsize=(8, 6))
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df[feature_to_plot], kde=True, color="blue")
plt.title(f"Histogram of{feature_to_plot}in Population growth in India")
plt.xlabel(feature_to_plot)
plt.ylabel("Frequency")
plt.show()
```
